## Constant Ice Density

```python
### set up our z-depths:
z_depths = np.arange(0, 100, 1)   # meters

## set c: 
c = 1000 # m
g = 9.8 # m/s^2
pw = 997 # kg/ m^3
pi = 917 # kg/ m^3
```

```python
## Approx. Solution [no shape function]
approx_ns = lambda z: ((-g*(pw - pi)*(c**2 - z**2)**(1/2))+784000)*((2/np.pi)*(c**0.5))
k1_ns = approx_ns(z_depths) * 10**-5  ## Convert to Mpa
```

```python
## Smith Solution [shape function]
approx_s = lambda z: 0.683*pw*(z**(3/2)) - 0.683*pi*(z**(3/2))
k1_s = approx_s(z_depths) * 10**-5 ## convert to Mpa
```

### plot

```python
plt.plot(k1_ns, z_depths, label = "Approx. Soln (no shape fct)")
plt.plot(k1_s, z_depths, label = "Smith Soln (shape fct)")
plt.legend()
plt.gca().invert_yaxis()
plt.ylabel("Crevasse Depth (m)")
plt.xlabel("K1 (Mpa)")
plt.title("Stress Intensity Factor")

### add a vertical line: 
plt.axvline(x=0.1, color='red', label = 'Critical Stress Intensity Factor')
```

## Discussion: 
**Is ignoring the shape factor justified?**
It seems like ignoring the shape factor is a good approximate fit, and estimates within the order of magnitude as well as general trend. As the crevasse is deeper, the more the shape factor matters. 

**What depth of water is required for the crack to propgate to the base of the glacier**
The depth of water required for the crack to propogate is approximately 31m.
